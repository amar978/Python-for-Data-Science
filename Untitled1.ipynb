# Distribution Plots

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
asd=sns.load_dataset('tips')
```

```python
asd.head(10)
```

```python
asd.shape
```

```python
sns.distplot(asd['total_bill'])
```

```python
sns.distplot(asd['total_bill'],kde=False,bins=30)
plt.grid()
```

```python
sns.jointplot(x='total_bill',y='tip',data=asd)
```

```python
sns.jointplot(x='total_bill',y='tip',data=asd,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=asd,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=asd,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=asd,kind='kde')
```

```python
sns.pairplot(asd)
```

```python
sns.pairplot(asd,hue='sex')
```

```python
sns.rugplot(asd['total_bill'])
```
